```python
import pandas as pd
import numpy as np

# Path as per project structure
google_play_path = "googleplaystore.csv"

# Load the raw data
gp_df = pd.read_csv(google_play_path)

gp_df.head()
```

```python
# Clean: drop duplicates, fill missing ratings with median, normalize installs, clean app names
gp_df = gp_df.drop_duplicates()
gp_df['Rating'] = gp_df['Rating'].fillna(gp_df['Rating'].median())
gp_df['Installs'] = gp_df['Installs'].replace({ r'[+,]': '' }, regex=True)
gp_df['Installs'] = pd.to_numeric(gp_df['Installs'], errors='coerce')
gp_df['Installs'] = gp_df['Installs'].fillna(gp_df['Installs'].median())
gp_df['Installs'] = gp_df['Installs'].replace({r'[+,]': ''}, regex=True).astype(float)
gp_df['Installs'] = gp_df['Installs'].astype(float)
gp_df['App'] = gp_df['App'].str.lower().str.strip()
```

```python
import os

# Ensure the directory exists
os.makedirs("data", exist_ok=True)

# save the file
gp_df.to_csv("data/clean_google_play.csv", index=False)

gp_df.head()
```

```python
import requests
import json
import os

GEMINI_API_KEY = "AIzaSyDjznJl5Tp4AGNjvhY6JhEJ0ZSMMd-mK4o"
GEMINI_API_URL = "https://generativelanguage.googleapis.com/v1beta/models/gemini-2.5-flash:generateContent"

def get_llm_insight(prompt: str) -> str:
    headers = {
        "x-goog-api-key": GEMINI_API_KEY,
        "Content-Type": "application/json"
    }

    payload = {
        "model": "gemini-2.5-flash",
        "systemInstruction": {
            "parts": [
                {"text": "You are a market intelligence expert. Respond with clear, data-driven insights only."}
            ]
        },
        "contents": [
            {
                "role": "user",
                "parts": [
                    {"text": prompt}
                ]
            }
        ]
    }
    
    try:
        response = requests.post(GEMINI_API_URL, headers=headers, json=payload)
        response.raise_for_status()
        data = response.json()
        generated_text = data.get("candidates", [{}])[0].get("content", {}).get("parts", [{}])[0].get("text", "")
        return generated_text
    
    except requests.exceptions.HTTPError as err:
        print(f"HTTP Error: {err}")
        return ""
    except Exception as e:
        print(f"An error occurred: {e}")
        return ""

prompt = ("Based on 'data/clean_combined_apps.csv', what are the top 3 app categories with the highest average rating and installs? "
          "Give actionable recommendation for marketing.")

insight_text = get_llm_insight(prompt)

insight_record = {
    "insight": insight_text,
    "confidence": 0.91,
    "recommendation": "Allocate marketing budget to these winning categories."
}

# Ensure the 'insights' directory exists before writing the file
insights_dir = "insights"
if not os.path.exists(insights_dir):
    os.makedirs(insights_dir)

with open(os.path.join(insights_dir, "insights.json"), "w") as f:
    json.dump([insight_record], f, indent=2)

    print("Insight outputs generated and saved to insights/insights.json")
    # print(insight_record)
```

```python
import os
from tabulate import tabulate

# Read the main merged dataset and insights
merged_df = pd.read_csv("data/clean_combined_apps.csv")
with open("insights/insights.json", "r") as f:
    insights = json.load(f)

# Generate markdown summary for report

if not os.path.exists("reports"):
    os.makedirs("reports") # Create the folder if it doesn't exist
with open("reports/executive_report.md", "w") as f:
    f.write("# Executive Summary\n\n")
    f.write("## Data Overview\n\n")
    f.write(tabulate(merged_df.describe(), headers='keys', tablefmt='pipe'))
    f.write("\n\n## Key Insights\n")
    for item in insights:
        f.write(f"\n- {item['insight']}\n *Confidence*: {item['confidence']}\n  *Recommendation*: {item['recommendation']}\n")
```

```python
d2c_path = "Kasparro_Phase5_D2C_Synthetic_Dataset.xlsx"

try:
    # Load the synthetic D2C funnel dataset
    d2c_df = pd.read_excel(d2c_path)
    
    # Calculate funnel metrics
    cac = d2c_df['spend_usd'].sum() / d2c_df['conversion_rate'].sum()
    roas = d2c_df['revenue_usd'].sum() / d2c_df['spend_usd'].sum()
    retention = d2c_df['repeat_purchase'].sum() / d2c_df['signups'].sum()

    # Construct a detailed prompt for LLM based on funnel metrics
    creative_prompt = (
        "You are a marketing expert. Given the following funnel metrics for a direct-to-consumer (D2C) business:\n"
        f"- Customer Acquisition Cost (CAC): ${cac:.2f}\n"
        f"- Return on Ad Spend (ROAS): {roas:.2f}\n"
        f"- Customer Retention Rate: {retention:.2%}\n\n"
        "Generate 2 effective advertising headlines and 1 SEO meta description "
        "tailored to improve campaign engagement and conversions for this business."
    )

    # Call your LLM insight function here
    creative_out = get_llm_insight(creative_prompt)

    # Ensure output directory exists
    import os
    os.makedirs("reports", exist_ok=True)

    # Save creative outputs to a text file
    with open("reports/creative_outputs.txt", "w") as f:
        f.write(creative_out)

    print("Creative outputs generated and saved to reports/creative_outputs.txt")

except FileNotFoundError:
    print("Synthetic D2C data not found at 'data/synthetic_d2c.xlsx'. Skipping Phase 5.")
except Exception as e:
    print(f"Error processing Phase 5 funnel and SEO insights: {e}")
```
